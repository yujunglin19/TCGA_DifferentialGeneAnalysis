```python
genes = [
    "IL13RA1","IL13RA2", "IL4", "IL4R","NOTCH1", "TOP2A", "CD44", "PTPRC", "CDK4", "HNRNPU", "PDGFRA", "CLCF1", "CCR1",
    "TGFBR2", "MYC", "TCF3", "CENPA", "SMC4", "CDC20", "CDC6", "SSRP1", "CHEK2", "SPHK1",
    "PTBP1", "SMC3", "DKC1", "POLR2D", "NIFK", "PMAIP1", "MECOM", "RBBP4", "SYNCRIP",
    "YBX1", "DDB2", "TDG", "WRAP53", "ELAVL1", "UBE2I", "SFN", "IKZF1", "ZEB2", "FAS",
    "NT5E", "IKBKB", "HLA-C", "VAV2", "RFT1", "SOD3", "RUNX1", "MCAM", "NES", "FKBP5",
    "IDH1", "LMNA", "ETV1", "FUS", "CHD1", "SCRIB", "TLN1", "FLNC", "GART", "NRXN1",
    "GAD2", "KCNC1", "CACNA1D", "SCN8A", "ATP2B2", "PPP3CA", "CAMK2B", "CALM3", "GABRA5",
    "VAMP2", "STK38", "LPL", "HLA-DPA1", "HAVCR2", "MYOF", "AQR", "EHD4", "AIMP1", "SNRNP40",
    "SSBP1", "MCPH1", "TMEM67", "CNTLN", "TTC25", "CEP152", "CC2D2A", "CLN3", "PGD", "SHOX2",
    "SMC1A", "LOXL2", "SPARC", "PDPN", "AGRN", "FBN3", "SEC24D", "KLRK1", "HK2", "TFEC",
    "EVI2B", "MED20", "SAMSN1" , "FMOD", "CCL2", "OSM", "CXCL13", "MNDA", "HLA-DRA"]
     #
```

```python
gene_dict = dict()
for ind in gene_mean_df.index:
  g = gene_mean_df['gene_name'][ind]
  m = gene_mean_df['mean'][ind]
  gene_dict[g] = m
```

```python
log_gene = []
for g in genes:
  m = gene_dict[g]
  val = list(GBM_df.loc[:, g])
  r = list(map(lambda x: math.log2((x+0.001)/m), val))
  n = 'log2_%s' % g
  log_gene.append(n)
  GBM_df[n] = r
```

```python
col = ['File Name', 'Case_ID', 'Sample ID', 'Sample Type', 'age_at_index',	'days_to_death']
new_col = col + [item for pair in zip(genes, log_gene + [0]) for item in pair]
new_col[:10]
```

```python
GBM_df = GBM_df.loc[:, new_col]
```
